# Plot boxplot and violin plot of cell-wise kNN Jaccard dissimilarity
from a per-cell CSV file.

Author: Tengxiao Gao

Description:
This script reads the per-cell dissimilarity results from
`RGC_scvi_lineage_vs_harmony_lineage_k50_to_2000_percell_k500.csv`
and generates two distribution plots:

1. Boxplot
2. Violin plot

The purpose is to visualize the distribution of cell-wise kNN
Jaccard dissimilarity for the comparison:
scvi_lineage vs harmony_lineage at k = 500.

Outputs:
- RGC_dissimilarity_box_k500.png
- RGC_dissimilarity_violin_k500.png

Input:
- CSV file containing at least the column `dissimilarity`

Usage:
Update the CSV path if needed, then run the script directly.

Notes:
- This script is for distribution-level visualization only
- It helps show spread, density, and outliers in dissimilarity values

```python
import pandas as pd
import matplotlib.pyplot as plt


# =========================================================
# 1. paths
# =========================================================
CSV_PATH = "RGC_scvi_lineage_vs_harmony_lineage_k50_to_2000_percell_k500.csv"

BOX_OUT = "RGC_dissimilarity_box_k500.png"
VIOLIN_OUT = "RGC_dissimilarity_violin_k500.png"


# =========================================================
# 2. load data
# =========================================================
df = pd.read_csv(CSV_PATH)

if "dissimilarity" not in df.columns:
    raise ValueError("CSV file must contain a 'dissimilarity' column.")

values = df["dissimilarity"].dropna()

print("[INFO] Number of cells:", len(values))
print("[INFO] Mean dissimilarity:", values.mean())
print("[INFO] Median dissimilarity:", values.median())


# =========================================================
# 3. boxplot
# =========================================================
plt.figure(figsize=(8, 7), dpi=200)

plt.boxplot(
    values,
    patch_artist=True,
    boxprops=dict(facecolor="lightgreen", color="gray", linewidth=1.5),
    medianprops=dict(color="gray", linewidth=1.5),
    whiskerprops=dict(color="gray", linewidth=1.5),
    capprops=dict(color="gray", linewidth=1.5),
    flierprops=dict(
        marker="o",
        markerfacecolor="white",
        markeredgecolor="gray",
        markersize=5,
        linestyle="none",
    ),
)

plt.ylabel("Jaccard dissimilarity", fontsize=16)
plt.title(
    "Cell-wise kNN dissimilarity (k = 500)\nscvi_lineage vs harmony_lineage",
    fontsize=20,
    pad=12,
)

plt.xticks([1], [""])
plt.ylim(0.3, 1.03)
plt.tight_layout()
plt.savefig(BOX_OUT, dpi=300, bbox_inches="tight")
plt.show()

print("[INFO] Saved:", BOX_OUT)


# =========================================================
# 4. violin plot
# =========================================================
plt.figure(figsize=(8, 7), dpi=200)

vp = plt.violinplot(
    values,
    showmeans=False,
    showmedians=True,
    showextrema=False,
)

for body in vp["bodies"]:
    body.set_facecolor("skyblue")
    body.set_edgecolor("gray")
    body.set_linewidth(1.5)
    body.set_alpha(1.0)

vp["cmedians"].set_color("gray")
vp["cmedians"].set_linewidth(1.5)

# add quartile lines manually
q1 = values.quantile(0.25)
q2 = values.quantile(0.50)
q3 = values.quantile(0.75)

plt.hlines([q1, q2, q3], xmin=0.85, xmax=1.15, colors="gray", linestyles="dashed", linewidth=1.2)

plt.ylabel("Jaccard dissimilarity", fontsize=16)
plt.title(
    "Cell-wise kNN dissimilarity (k = 500)\nscvi_lineage vs harmony_lineage",
    fontsize=20,
    pad=12,
)

plt.xticks([1], [""])
plt.ylim(0.3, 1.03)
plt.tight_layout()
plt.savefig(VIOLIN_OUT, dpi=300, bbox_inches="tight")
plt.show()

print("[INFO] Saved:", VIOLIN_OUT)
```
