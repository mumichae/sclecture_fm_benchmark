## Download complete dataset (3M+ cells)
Consider that the whole dataset is 35GB! Only download the complete dataset when and where you really need it.
For testing and developing any of the subsets (see table above) should be sufficient.

```python
url = 'https://datasets.cellxgene.cziscience.com/b4b58102-d03b-47b9-93b1-ed419b042de7.h5ad'
file_path = 'cxg/retina_all.h5ad'

!curl -C - -o {file_path} {url}

# note that -C resumes incomplete downloads and skips the download entirely, 
# if the dataset has already been downloaded
```

Load data lazily with [`dask`](https://anndata.readthedocs.io/en/latest/tutorials/notebooks/anndata_dask_array.html)

```python
adata = read_anndata('cxg/retina_all.h5ad', dask=True, backed=True)
adata
```
